Usuwam kolumny z najwieksza iloscią braków

```python
data = df.drop(['Event.Id', 'Latitude', 'Longitude', 'Airport.Code', 'Airport.Name', 'Aircraft.Category', 'Registration.Number', 'FAR.Description', 'Schedule', 'Air.carrier', 'Report.Status', 'Publication.Date'],  axis=1)
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
sns.heatmap(data.corr())
```

```python
data = data.dropna(subset=('Country', 
                           'Location','Aircraft.damage', 'Make', 'Total.Fatal.Injuries', 'Total.Serious.Injuries', 
                           'Total.Minor.Injuries', 'Total.Uninjured', 'Broad.phase.of.flight') ).reset_index() #usuwam puste wiersze w wazniejszych kolumnach 
```

```python
data.isnull().sum()
```

```python
braki = data[data.isna().any(axis=1)] #df tylko z wierszami, ktore zawierają braki
braki
```

Uzupełniam puste wiersze - mode()

```python
data['Purpose.of.flight'] = data['Purpose.of.flight'].fillna(data['Purpose.of.flight'].mode()[0])
data['Weather.Condition'] = data['Weather.Condition'].fillna(data['Weather.Condition'].mode()[0])
data['Engine.Type'] = data['Engine.Type'].fillna(data['Engine.Type'].mode()[0])
data['Model'] = data['Model'].fillna(data['Model'].mode()[0])
```

```python
data.isnull().sum()
```
