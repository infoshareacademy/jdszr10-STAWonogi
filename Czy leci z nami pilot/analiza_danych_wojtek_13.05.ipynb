# finalnie wyszła mi tabela w której są : liczba ofiar śmiertelnych, liczba osób bez obrażeń (niekoniecznie potrzebna), warunki pogodowe, fazy lotu, uszkodzenia samolotu. z ok 88 tys wierszy bez danych dot. fazy lotu jest 27 165 wierszy (vide ostatnia komórka)  - to wypadałoby odrzucić, kolumnę fatal.injuries zamieniłbym na 1 / 0 (były  / nie były), i analizować możnaby pod kątem która faza lotu jest najbardziej ryzykowna jeśli chodzi o utratę życia ?? 

```python
wypadki_4 = pd.get_dummies(wypadki_x)
```

```python
wypadki_4.head()
```

```python
wypadki_4.describe()
```

```python
wypadki_4.info()
```

```python
wypadki_4['Total.Fatal.Injuries'] = wypadki_4['Total.Fatal.Injuries'].apply(lambda x: x>0)
#df['age'] = df['age'].apply(lambda x: x > 18)
```

```python
wypadki_4['Total.Fatal.Injuries'] = wypadki_4['Total.Fatal.Injuries'].astype(int)
```

```python
wypadki_4.info()
```

```python
wypadki_4.head()
```

```python
features = wypadki_4.iloc[:, 2:21]
```

```python
X = features.values
y = wypadki_4['Total.Fatal.Injuries'].values
print(X.shape)
print(y.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(X_train, y_train)
```

```python
logreg.score(X_train, y_train)
```

```python
logreg.score(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
plt.figure(figsize = (12, 8))

plot_tree(model, feature_names = wypadki_4.columns,
          class_names = wypadki_4.columns,
          filled=True);
```

```python
model.feature_importances_
```

```python
y_axis = list(wypadki_4.columns[2:21])
x_axis = model.feature_importances_
plt.barh(y_axis, x_axis)
plt.show()
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
